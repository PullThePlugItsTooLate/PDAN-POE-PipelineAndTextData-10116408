## Theory

### What is Text Processing?

Text processing is a critical aspect of data analysis, involving the transformation of raw text into a format that can be efficiently analyzed by computers. Given that text data is often unstructured, text processing encompasses various techniques to clean, structure, and convert this data into a machine-readable format. This conversion is essential for extracting meaningful patterns and insights from text, enabling a range of applications from classification to topic modeling (jacobperalta, 2024).

#### Process

Text processing begins with tokenization, the process of breaking down text into individual units called tokens, which can be words, phrases, or symbols. For example, the sentence "Text processing is important" can be tokenized into ["Text", "processing", "is", "important"]. This step simplifies the text, making it easier to analyze and manipulate (Deepanshi, 2024).

Next, stop words are typically removed, which are commonly used words that carry little semantic value, such as "is", "and", and "the". By removing stop words, we reduce noise in the data, allowing us to focus on more significant words. For instance, the sentence "The quick brown fox jumps over the lazy dog" becomes ["quick", "brown", "fox", "jumps", "lazy", "dog"] after stop words are removed (Deepanshi, 2024).

Stemming is another key technique, where words are reduced to their root or base form. For example, "running", "runner", and "ran" can all be stemmed to "run". This step helps in grouping similar words together, thus reducing the complexity of the text and improving the performance of text analysis algorithms (Deepanshi, 2024).

Lemmatization is similar to stemming but more advanced, as it reduces words to their base or dictionary form, considering the morphological analysis of the words. For example, "better" can be lemmatized to "good". Lemmatization helps in standardizing words and maintaining meaningful forms, which is crucial for tasks that require precise word meanings (Deepanshi, 2024).

The Term Frequency-Inverse Document Frequency (TF-IDF) is a statistical measure used to evaluate the importance of a word in a document relative to a collection of documents. It is calculated by multiplying term frequency (TF) by inverse document frequency (IDF). TF-IDF helps in identifying the most relevant words in a text corpus by highlighting words that are frequent in a document but rare across the corpus, making it crucial for feature extraction in text mining (Deepanshi, 2024).

Latent Dirichlet Allocation (LDA) is a topic modeling technique that discovers hidden topics in a set of documents by grouping words that frequently occur together. LDA helps in uncovering the underlying structure in text data, making it useful for tasks like topic extraction and content summarization (Kulshrestha, 2019).

#### Need for Text Processing

Text processing is essential for handling unstructured data, which is often noisy and difficult to analyze directly. By structuring the data, text processing facilitates more effective and meaningful analysis, improving the accuracy of algorithms in identifying patterns and trends. Techniques like TF-IDF and LDA enable sophisticated analyses such as keyword extraction and topic modeling, providing deeper insights into the text data (DataRobot, 2020).

#### Applications of Text Processing

Text processing has a wide range of applications. It is used in sentiment analysis to identify the sentiment expressed in text, which is valuable for understanding public opinion or customer feedback. It also plays a crucial role in document classification, where documents are categorized into predefined categories based on their content, such as classifying news articles into topics like sports or politics. Furthermore, text processing enhances information retrieval by improving the relevance of search results through the analysis of document and query content (DataRobot, 2020).

#### In Summation

All in all, text processing transforms raw text into a format suitable for analysis, enabling the extraction of valuable insights. This process is vital for any task involving the analysis of text data, from simple keyword extraction to complex topic modeling.

The `cleaner()` function performs several text preprocessing tasks, including replacing non-alphabetic characters with spaces, removing digits, condensing repeated characters to a single character, and removing URLs. The cleaned text is then applied to the 'Tweet' column of the data DataFrame, resulting in a dataset with tweets that have been preprocessed.

```python
# Celius Stingher
# https://stackoverflow.com/questions/64719706/cleaning-twitter-data-pandas-python

def cleaner(text):
    text = re.sub('[^a-zA-Z]', ' ', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1+', r'\1', text)
    text = re.sub(r'http[s]?://\S+', '', text)
    text = re.sub(r'www\.\S+', '', text)
    return text

data['Tweet'] = data['Tweet'].apply(cleaner)
data
```

### Text Preprocessing

Next up we're going to set up the necessary resources for text preprocessing, define stop words, initialize a lemmatizer, and create a function to preprocess text data. The function tokenizes the text, removes stop words and non-alphabetic tokens, applies lemmatization, and joins the processed tokens back into a single string. Finally, it applies this preprocessing function to the tweets in the dataset and adds the results as a new column, displaying the first few rows of the updated dataset so we can check out the changes.

```python
# Tithi Sreemany
# https://www.analyticsvidhya.com/blog/2021/09/essential-text-pre-processing-techniques-for-nlp/

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    tokens = nltk.word_tokenize(text)
    tokens = [word for word in tokens if word.lower() not in stop_words and word.isalpha()]
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(lemmatized_tokens)

data['processed_text'] = data['Tweet'].astype(str).apply(preprocess_text)
data.head()
```

Next, to make data easier to process, we will use a Label Encoder - essentially changing the Normal tweets to 0 and the Suicidal ones to 1

```python
# Moez Ali
# https://www.datacamp.com/tutorial/categorical-data

encoder = LabelEncoder()
data['Label'] = encoder.fit_transform(data['Label'])
data.head()
```

### Pipeline Integration

We will define a custom transformer class called `TextPreprocessor` for use in a scikit-learn pipeline. This class inherits from TransformerMixin, which is a mixin class that provides the necessary methods to work seamlessly with scikit-learn's pipeline.

- `transform`: Applies the preprocess_text function to each text document in the input data.
- `fit`: Does nothing and just returns the instance itself, as no fitting is needed for this preprocessing step. (Although useful to keep around for future extensions)

```python
# jacobperalta
# https://www.geeksforgeeks.org/text-preprocessing-in-python-set-1/

class TextPreprocessor(TransformerMixin):
    def transform(self, X, **transform_params):
        return [preprocess_text(text) for text in X]
    
    def fit(self, X, y=None, **fit_params):
        return self
```

Next lets write a block that allows the user to select a machine learning classifier for a text classification task and then run the selected classifier through a pipeline that includes text preprocessing, TF-IDF vectorization, and the chosen classifier. This way you could just enter a number in to choose which model you wish to test out and you can even add more models to the if else statement in order to examine more.

The pipeline consists of three steps:
- `preprocessor`: An instance of the TextPreprocessor class, which preprocesses the text data.
- `tfidf`: An instance of TfidfVectorizer, which converts the preprocessed text into TF-IDF features.
- `classifier`: The classifier selected based on user input.

I've done the training and testing splits in the following way:
- `test_size=0.2`: 20% of the data is used for testing.
- `random_state=42`: Ensures reproducibility of the results.


```python
# Asjad Ali
# https://medium.com/@asjad_ali/understanding-the-nlp-pipeline-a-comprehensive-guide-828b2b3cd4e2

# Nilay Chauhan
# https://www.kaggle.com/code/nilaychauhan/getting-started-with-nlp-pipelines

classifierSelection = input('Enter classification model (1: RandomForest, 2: XGBoost, 3: Naive Bayes): ')

if classifierSelection == "1":
    classifier = RandomForestClassifier()
elif classifierSelection == "2":
    classifier = XGBClassifier()
elif classifierSelection == "3":
    classifier = MultinomialNB()    
else:
    print("You have not made a valid selection")
    classifier = None

if classifier is not None:
    pipeline = Pipeline([
        ('preprocessor', TextPreprocessor()),
        ('tfidf', TfidfVectorizer(max_features=1000)),
        ('classifier', classifier)
    ])

    X_train, X_test, y_train, y_test = train_test_split(data['processed_text'], data['Label'], test_size=0.2, random_state=42)
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    print(f'Pipeline Accuracy: {accuracy * 100:.2f}%')
    print(classification_report(y_test, y_pred))

    conf_matrix = confusion_matrix(y_test, y_pred)
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=encoder.classes_, yticklabels=pipeline.classes_)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()
else:
    print("Please enter either 1, 2, or 3 as a classifier. Classifier mapping available in code")
```
